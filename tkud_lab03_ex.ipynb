### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

````python
plt.plot(np.sin(np.linspace(0,2*np.pi,180)),label="sin(x)")
plt.plot(np.cos(np.linspace(0,2*np.pi,180)),label="cos(x)",ls='--')

plt.legend()
plt.show()
````

### Câu 5: Vẽ các đồ thị con

````python
plt.subplot(2, 1, 1)
plt.plot(np.sin(np.linspace(0,2*np.pi,180)),label="sin(x)")
plt.legend()


plt.subplot(2, 1, 2)
plt.plot(np.cos(np.linspace(0,2*np.pi,180)),'r',label="cos(x)")
plt.legend()



plt.show()
````
